```python
## Imports
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy.optimize import curve_fit
import time as t
```

```python
## This function should be an exact one-to-one with the MATLAB
def load_tiff_movie(pathname, filename):
    ftbo = pathname + filename
    info = cv2.imread(ftbo, 0) # read first frame to get dimensions
    numframe = len(info)
    rawframes = []
    for ii in range(numframe):
        rawframes.append(cv2.imread(ftbo, ii))
    return rawframes, numframe
```

```python
## Gaussian_image_filtering.m translated, should work the same after one tweak (noted below)
def Gaussian_image_filtering(I, R):
    # Compute FFT of the grey image
    A = np.fft.fft2(I.astype(float))
    A1 = np.fft.fftshift(A) # frequency scaling
    
    # Gaussian Filter Response Calculation
    M, N = I.shape
    X, Y = np.meshgrid(np.arange(N), np.arange(M))
    Cx, Cy = 0.5*N, 0.5*M
    ## Lo = np.exp(-((X-Cx)**2 + (Y-Cy)**2) / (2*R**2)) ## Original ChatGPT Translation
    ## Fixed below? I was comparing to the MATLAB and I think the 2 is squared as well in (2*R**2)
    Lo = np.exp(-((X-Cx)**2 + (Y-Cy)**2) / ((2*R)**2)) 
    Hi = 1 - Lo # High pass filter=1-low pass filter
    
    # Filtered image=ifft(filter response*fft(original image))
    J = A1 * Lo
    J1 = np.fft.ifftshift(J)
    B1 = np.fft.ifft2(J1).real
    K = A1 * Hi
    K1 = np.fft.ifftshift(K)
    B2 = np.fft.ifft2(K1).real
    
    # Display input image and filtered output
    fig, axs = plt.subplots(1, 2)
    axs[0].imshow(I, cmap='gray')
    axs[0].set_title('Input image')
    axs[0].axis('off')
    axs[1].imshow(np.abs(B1), cmap='gray')
    axs[1].set_title('Filtered image')
    axs[1].axis('off')
    plt.show()
    
    return B1, B2
```

```python
## There are two versions of this with a slight change in between, I think the second one works better but still not sure
def FastPeakFind(d, thres=None, filt=None, edg=3, res=1, fid=None):
    """
    Analyze noisy 2D images and find peaks using local maxima (1 pixel
    resolution) or weighted centroids (sub-pixel resolution).
    The code is designed to be as fast as possible, so I kept it pretty basic.
    The code assumes that the peaks are relatively sparse, test whether there
    is too much pile up and set threshold or user defined filter accordingly.
    """
    # defaults
    original = d.copy()
    
    if d.ndim > 2: # added in case one uses imread (JPG\PNG\...).
        d = np.uint16(np.mean(d, axis=-1))
        
    if np.issubdtype(d.dtype, np.floating):
        if np.max(d) <= 1:
            d = np.uint16(d * (2**16 / np.max(d)))
        else:
            d = np.uint16(d)
    
    if thres is None:
        thres = np.min([np.max(d, axis=0).min(), np.max(d, axis=1).min()])
    
    if filt is None:
        filt = np.zeros((7,7))
        filt[3,3] = 1
        filt = ndimage.gaussian_filter(filt, 1)
    
    # analyze image
    d = ndimage.median_filter(d, size=3)
    d = ndimage.convolve(d, filt, mode='constant', cval=0.0)
    
    # apply threshold
    d = d * (d > thres)
    
    # peak finding
    if res == 1:
        peaks = ndimage.maximum_filter(d, size=3, mode='constant', cval=0)
        mask = (d == peaks)
        y, x = np.nonzero(mask)
    elif res == 2:
        peaks, num = ndimage.label(d > thres)
        y, x = ndimage.center_of_mass(d, peaks, range(1, num+1))
    else:
        raise ValueError("Invalid res value. Must be 1 or 2.")
    
    # remove edge peaks
    mask = (x > edg) & (x < (d.shape[1]-edg)) & (y > edg) & (y < (d.shape[0]-edg))
    x = x[mask]
    y = y[mask]
    
    # output results
    cent = np.concatenate((x.reshape(-1, 1), y.reshape(-1, 1)), axis=1).flatten()
    
    ##############################################################################
    if res == 1:
        cm = np.zeros_like(d)
        cm[mask] = 1
        varargout = (cm,)
    else:
        varargout = ()
    ##############################################################################
    
    if fid is not None:
        np.savetxt(fid, cent, fmt='%d', delimiter=',')
    
    return cent, *varargout
```

```python
## Slightly different version, some changes in between the comment lines
def FastPeakFind(d, thres=None, filt=None, edg=3, res=1, fid=None):
    """
    Analyze noisy 2D images and find peaks using local maxima (1 pixel
    resolution) or weighted centroids (sub-pixel resolution).
    The code is designed to be as fast as possible, so I kept it pretty basic.
    The code assumes that the peaks are relatively sparse, test whether there
    is too much pile up and set threshold or user defined filter accordingly.
    """
    # defaults
    original = d.copy()
    
    if d.ndim > 2: # added in case one uses imread (JPG\PNG\...).
        d = np.uint16(np.mean(d, axis=-1))
        
    if np.issubdtype(d.dtype, np.floating):
        if np.max(d) <= 1:
            d = np.uint16(d * (2**16 / np.max(d)))
        else:
            d = np.uint16(d)
    
    if thres is None:
        thres = np.min([np.max(d, axis=0).min(), np.max(d, axis=1).min()])
    
    if filt is None:
        filt = np.zeros((7,7))
        filt[3,3] = 1
        filt = ndimage.gaussian_filter(filt, 1)
    
    # analyze image
    d = ndimage.median_filter(d, size=3)
    d = ndimage.convolve(d, filt, mode='constant', cval=0.0)
    
    # apply threshold
    d = d * (d > thres)
    
    # peak finding
    if res == 1:
        peaks = ndimage.maximum_filter(d, size=3, mode='constant', cval=0)
        mask = (d == peaks)
        y, x = np.nonzero(mask)
    elif res == 2:
        peaks, num = ndimage.label(d > thres)
        y, x = ndimage.center_of_mass(d, peaks, range(1, num+1))
    else:
        raise ValueError("Invalid res value. Must be 1 or 2.")
    
    # remove edge peaks
    mask = (x > edg) & (x < (d.shape[1]-edg)) & (y > edg) & (y < (d.shape[0]-edg))
    x = x[mask]
    y = y[mask]
    
    # output results
    cent = np.concatenate((x.reshape(-1, 1), y.reshape(-1, 1)), axis=1)
    
    ##############################################################################
    if res == 1:
        cm = np.zeros(d.shape, dtype=bool)
        cm[x, y] = True
        varargout = (cm,)
    else:
        varargout = ()
    ##############################################################################
    
    if fid is not None:
        np.savetxt(fid, cent, fmt='%d', delimiter=',')
    
    return cent, *varargout
```

```python
## These are two helper functions ChatGPT wrote for itself for the code block after this 
## the PSF function will need some investigating because it is trying to replicate a MATLAB function on its own
## and I'm not sure if this function is going to do the right thing.
def gauss_2d(x, y, x0, y0, A, sigma_x, sigma_y):
    return A * np.exp(-((x-x0)**2/(2*sigma_x**2) + (y-y0)**2/(2*sigma_y**2)))

def PSFfit(MOLE, pxl_size):
    N = MOLE.size
    x, y = np.meshgrid(np.arange(MOLE.shape[1]), np.arange(MOLE.shape[0]))

    x0 = np.sum(MOLE * x) / np.sum(MOLE)
    y0 = np.sum(MOLE * y) / np.sum(MOLE)

    p0 = [x0, y0, MOLE.max(), 1, 1]
    bounds = ([0, 0, 0, 0, 0], [MOLE.shape[1], MOLE.shape[0], np.inf, np.inf, np.inf])

    try:
        popt, pcov = curve_fit(gauss_2d, (x, y), MOLE.ravel(), p0=p0, bounds=bounds)
    except:
        return np.nan, np.nan, N

    sig_x = popt[3] * pxl_size
    sig_y = popt[4] * pxl_size

    return sig_x, sig_y, N
```

```python
import numpy as np
from gaussfitter import gaussfit as gauss2dfit

def PSFfit(IMAGE, pxl_size):
    ny, nx = abs(IMAGE).shape

    grid = np.concatenate((np.array([nx, ny]), np.arange(1, nx+1), np.arange(1, ny+1)))
    popt, _, _, _ = gauss2dfit(abs(IMAGE), grid, returnfitimage=False)

    if len(popt) != 0:
        xp = np.reshape(np.repeat((grid[2:nx+2] - popt[4])/popt[2], ny), ny*nx)
        yp = np.reshape(np.repeat((grid[nx+2:] - popt[5])/popt[3], nx), ny*nx)
        expU = np.exp(-0.5 * (xp**2 + yp**2))
        z = popt[0] + popt[1] * expU

        N = np.sum(z - popt[0])
        z = np.reshape(z, (ny, nx))

        si_x = popt[2] * pxl_size
        si_y = popt[3] * pxl_size
        a = pxl_size

        bkg = np.reshape(IMAGE - z, -1)
        b = np.std(bkg)

        sig_ax2 = si_x**2 + a**2/12
        sig_ay2 = si_y**2 + a**2/12

        sig_x = np.sqrt(sig_ax2/N * (16/9 + 8*np.pi*sig_ax2*b**2/(N*a**2)))
        sig_y = np.sqrt(sig_ay2/N * (16/9 + 8*np.pi*sig_ay2*b**2/(N*a**2)))

    else:
        sig_x = np.array([])
        sig_y = np.array([])
        N = np.array([])

    return sig_x, sig_y, N
```
